```python
# Handle missing values in train_df
for col in train_df.columns:
    if train_df[col].isnull().any():
        if train_df[col].dtype == 'object':
            train_df[col] = train_df[col].fillna(train_df[col].mode()[0])
        else:
            train_df[col] = train_df[col].fillna(train_df[col].mean())

# Handle missing values in test_df (using the mode/mean from the training data)
for col in test_df.columns:
    if test_df[col].isnull().any():
        if test_df[col].dtype == 'object':
            test_df[col] = test_df[col].fillna(train_df[col].mode()[0]) # Use train_df mode
        else:
            test_df[col] = test_df[col].fillna(train_df[col].mean()) # Use train_df mean

print("Missing values in train_df after handling:")
display(train_df.isnull().sum())

print("\nMissing values in test_df after handling:")
display(test_df.isnull().sum())
```

```python
# Feature Engineering

# Create a new feature: TotalCharges per AccountAge
train_df['ChargesPerMonth'] = train_df['TotalCharges'] / train_df['AccountAge']
test_df['ChargesPerMonth'] = test_df['TotalCharges'] / test_df['AccountAge']

# You can add more feature engineering steps here
# Example: Interaction terms, polynomial features, etc.

print("Train_df with new feature:")
display(train_df[['AccountAge', 'TotalCharges', 'ChargesPerMonth']].head())

print("\nTest_df with new feature:")
display(test_df[['AccountAge', 'TotalCharges', 'ChargesPerMonth']].head())
```

```python
# Preprocessing Categorical Features

# Identify categorical columns (excluding CustomerID and the target variable Churn)
categorical_cols = train_df.select_dtypes(include='object').columns.tolist()
if 'CustomerID' in categorical_cols:
    categorical_cols.remove('CustomerID')
if 'Churn' in categorical_cols:
    categorical_cols.remove('Churn')

print(f"Categorical columns to encode: {categorical_cols}")

# Apply one-hot encoding to the training data
train_df_encoded = pd.get_dummies(train_df, columns=categorical_cols, drop_first=True)

# Apply one-hot encoding to the test data, ensuring consistent columns
test_df_encoded = pd.get_dummies(test_df, columns=categorical_cols, drop_first=True)

# Align columns - crucial for consistent feature sets between train and test
train_cols = list(train_df_encoded.columns)
test_cols = list(test_df_encoded.columns)

# Remove the target variable 'Churn' from train_cols for alignment
if 'Churn' in train_cols:
    train_cols.remove('Churn')

# Find columns present in train but not test, and add them to test_df_encoded with zeros
missing_in_test = set(train_cols) - set(test_cols)
for c in missing_in_test:
    test_df_encoded[c] = 0

# Find columns present in test but not train, and drop them from test_df_encoded
missing_in_train = set(test_cols) - set(train_cols)
for c in missing_in_train:
    test_df_encoded = test_df_encoded.drop(c, axis=1)

# Ensure the order of columns is the same
test_df_encoded = test_df_encoded[train_cols]

print("\nShape of train_df after one-hot encoding:", train_df_encoded.shape)
print("Shape of test_df after one-hot encoding:", test_df_encoded.shape)

display(train_df_encoded.head())
display(test_df_encoded.head())
```

```python
# Select features and target
X_train = train_df_encoded.drop(['CustomerID', 'Churn'], axis=1)
y_train = train_df_encoded['Churn']
X_test = test_df_encoded.drop('CustomerID', axis=1)

# Ensure columns match between training and testing features
# This step is crucial if columns were not perfectly aligned during one-hot encoding
# (although our previous alignment step should handle this)
# If X_train and X_test columns don't match, you'll need to re-align them here.
# For example:
# train_cols = X_train.columns
# test_cols = X_test.columns
# missing_in_test = set(train_cols) - set(test_cols)
# for c in missing_in_test:
#     X_test[c] = 0
# missing_in_train = set(test_cols) - set(train_cols)
# for c in missing_in_train:
#     X_test = X_test.drop(c, axis=1)
# X_test = X_test[train_cols]


# Select and Train a Model (Step 4 of the plan)
from sklearn.ensemble import RandomForestClassifier

# Initialize the model - you can tune hyperparameters later
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
# Convert y_train to integer type
model.fit(X_train, y_train.astype(int))

# Make Predictions (Step 5 of the plan)
# Predict probabilities
test_predictions_proba = model.predict_proba(X_test)[:, 1]

# Create Submission File (Step 6 of the plan)
# Create the prediction_df dataframe
prediction_df = pd.DataFrame({'CustomerID': test_df_encoded['CustomerID'],
                              'predicted_probability': test_predictions_proba})

# View our 'prediction_df' dataframe as required for submission.
# Ensure it should contain 104,480 rows and 2 columns 'CustomerID' and 'predicted_probaility'
print(prediction_df.shape)
display(prediction_df.head(10))
```
